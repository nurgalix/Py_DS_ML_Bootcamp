```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```
