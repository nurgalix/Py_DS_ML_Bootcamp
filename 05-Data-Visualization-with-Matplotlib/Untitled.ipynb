```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# OO
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.15,0.4, 0.5, 0.4])
axes1.set_title('Larger')
axes2.set_title('Smaller')
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(x,y)
axes[1].set_title('Second Plot')
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label = 'X Squared')
ax.plot(x, x ** 3, label = 'X Cubed')

ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=0.5, alpha=1, ls='-', marker='o', markersize=20,
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=2, ls='--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
